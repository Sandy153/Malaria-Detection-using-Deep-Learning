```python
numbers = input("Enter numbers separated by commas: ").split(",")
numbers = [int(num) for num in numbers] # Convert input to integers
print("Original List:", numbers)
```

```python
size = len(numbers)
print("Size of the list:", size)
```

```python
numbers.sort()
print("Sorted List:", numbers)
```

```python
mean = sum(numbers) / size
print("Mean:", mean)
```

```python
if size % 2 == 0:
          median = (numbers[size//2 - 1] + numbers[size//2]) / 2
else:

    median = numbers[size//2]
print("Median:", median)
```

```python
number_counts = {num: numbers.count(num) for num in numbers}
mode = max(number_counts, key=number_counts.get)
print("Mode:", mode)
```

```python
variance = sum((x - mean) ** 2 for x in numbers) / size
print("Variance:", variance)
```

```python
std_dev = variance ** 0.5
print("Standard Deviation:", std_dev)
```
