```python
import math # For basic mathematical operations
import statistics # For statistical functions
import numpy as np # For numerical operations on arrays
import scipy # For advanced scientific computations
```

```python
print(math.sqrt(16)) # Square root of 16
print(math.pow(2, 3)) # 2 raised to the power 3
print(math.log(10)) # Natural logarithm of 10
print(math.sin(math.radians(30))) # Sine of 30 degrees
```

```python
data = [10, 20, 30, 40, 50, 50]
print(statistics.mean(data)) # Mean
print(statistics.median(data)) # Median
print(statistics.mode(data)) # Mode
print(statistics.stdev(data)) # Standard Deviation
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(np.mean(arr)) # Mean using NumPy
print(np.median(arr)) # Median using NumPy
print(np.std(arr)) # Standard deviation
print(np.sin(np.pi / 2)) # Sine of 90 degrees (π/2 radians)
```
